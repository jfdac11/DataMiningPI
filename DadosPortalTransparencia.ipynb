#### RETIRANDO O '*' INÍCIO

```python
def limpeza(row):
#     stri = row['LINHAA']
    for c in col_cids:
        if row[c].find('*') != -1:
            row[c] = row[c][1:]
#             print(stri)
#         print(row.LINHAA[1:])
#     row = row.CODMUNOCOR[:2]

cids.apply(limpeza, axis='columns')
cids.head()
```

```python
cids['GRUPO_CAUSA_BASICA'] = ""
cids['GRUPO_CID_A'] = ""
cids['GRUPO_CID_B'] = ""
cids['GRUPO_CID_C'] = ""
cids['GRUPO_CID_D'] = ""
cids['CAUSA_BASICA'] = ""
cids['CID_A'] = ""
cids['CID_B'] = ""
cids['CID_C'] = ""
cids['CID_D'] = ""
```

```python
def agrupamento_cid(row):
    cids.loc[((cids.CAUSABAS.str[0] == row.LETRA_INICIO) | (cids.CAUSABAS.str[0] == row.LETRA_FIM)) 
             & ((cids.CAUSABAS.str[1:] >= row.NUM_INICIO) & (cids.CAUSABAS.str[1:] <= row.NUM_FIM)),
             ['GRUPO_CAUSA_BASICA']] = row.DESCRICAO 
df_cid_grupos.apply(agrupamento_cid, axis='columns')
```

```python
cids.head()
```

```python
def categoria_cid(row):
#     if len(row['']) < 3 
    cids.loc[((cids.CAUSABAS.str.len() == 3) & (cids.CAUSABAS == row.CAT)),
             ['CAUSA_BASICA']] = row.DESCRICAO 
#     cids.loc[((cids.CAUSABAS.str.len() == 4) & (cids.CAUSABAS == df_cid_sub.SUBCAT[0]))]
df_cid_cate.apply(categoria_cid, axis='columns')
```

```python
cids['CATEGORIA_BASICA'] = ""
```

```python
cids.head()
```

```python
def sub_categoria_cid(row):
#     print(row.DESCRICAO)
    cids.loc[((cids.CAUSABAS.str.len() == 4) & (cids.CAUSABAS == row.SUBCAT)),
            ['CAUSA_BASICA']] = row.DESCRICAO
df_cid_sub.apply(sub_categoria_cid, axis='columns')
```
