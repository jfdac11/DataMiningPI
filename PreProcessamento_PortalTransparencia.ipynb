#Funções para tradução dos dados

####Idades: 




##### ⏰
* Para 2 dígitos: menos de 1 ano de vida


* Para 3 dígitos com menor do que 400: menos de 1 ano de vida


* Para 3 dígitos maiores do que 400: 
  * 2 últimos significam o número de anos
  * o 1 primeiro significa  número de centenário a partir de 400


````python
def ajeitando_idades(df):
    idades = pd.DataFrame(df['IDADE']).astype('int')
#     print(idades.head())
    idades['ANOS_IDADE'] = pd.Series()
    
    idades.loc[idades['IDADE'] < 400, ['ANOS_IDADE']] = 0
    
    idades.loc[idades['IDADE'] > 400, ['ANOS_IDADE']] = (idades['IDADE'] - 400)
    
    return idades['ANOS_IDADE']
````

````python
def faixa_etaria(df):
    faixas_etarias = np.arange(0, df.IDADE.max()+1, 5, dtype='int')
    df['FAIXA_ETARIA'] = pd.cut(df['IDADE'], faixas_etarias)
````

````python
def add_zero(row, name):
    if len(row[name]) < 8:
        row[name] = '0' + row[name]
````

#### Corrigindo as datas, pois algumas não possuem 8 carácteres

````python
def clean_date(df, name_column):
    dt_nasc = pd.DataFrame(df[name_column])
    
    dt_nasc[name_column] = dt_nasc[name_column].fillna(0)
    dt_nasc[name_column] = dt_nasc[name_column].astype('int')
    dt_nasc[name_column] = dt_nasc[name_column].astype('str')
    

    dt_nasc.apply(add_zero, name=name_column, axis='columns')
    dt_nasc[name_column] = dt_nasc[name_column].str.slice(0, 8)
    
    
    dt_nasc['ANO'] = dt_nasc[name_column].str[4:]

    dt_nasc['MES'] = dt_nasc[name_column].str.slice(2, 4)

    dt_nasc['DIA'] = dt_nasc[name_column].str.slice(0, 2)
    

    return dt_nasc
````

#### Pegando o parte do código do município equivalente ao estado para atribuir sua respectiva sigla (UF)

````python
def codigos(row):
    if len(row['CODMUNOCOR']) < 7:
#         print('0' + row.DTOBITO[0])
        row['COD'] = row.CODMUNOCOR[:2]
````

##### Atribuindo siglas das UFs de acordo com seus códigos segundo o CSV dos estados do IBGE

````python
def clean_states(df):
    data_estados = pd.read_csv(path + '/CSVs BASE/estados_ibge.csv', sep=';', header=0, encoding='latin1')

    cod = pd.Series(data_estados['CODIGO'].astype('str'))
    cod = cod.values

    uf = pd.Series(data_estados['UF'].astype('str'))
    uf = uf.values

    cod_mun = pd.DataFrame(df.CODMUNOCOR.astype('str'))

    cod_mun['CODMUNOCOR'] = cod_mun.CODMUNOCOR.astype('str')
    cod_mun['CODMUNOCOR'].str.slice(0,2)
    cod_mun['COD'] = ""
    cod_mun.apply(codigos, axis='columns')
    cod_mun['ESTADO'] = cod_mun['COD'].replace(cod, uf)

    return cod_mun['ESTADO'], cod_mun['COD']
````

####Retirando * das tuplas em que a causa básica contém antes do valor CID

````python
def limpeza(row, col_cids):
    for c in col_cids:
        if row[c].find('*') != -1:
            row[c] = row[c][1:]
````

#####Agrupando causa básica de acordo com a tabela CAPÍTULOS do CID-10

````python
def capitulo_cid(row, cids):
    cids.loc[( ( (cids.CAUSABAS.str[0] == row.CATINIC[0]) | (cids.CAUSABAS.str[0] == row.LETRA_FIM) )
              &
               ((cids.CAUSABAS.str[1:3] >= row.CATINIC[1:]) & (cids.CAUSABAS.str[1:3] <= row.CATFIM[1:]))
             ),
             ['CAPITULO_CAUSA_BASICA', 'COD_CAPITULO_CAUSA_BASICA']] = [row.DESCRABREV, row.NUMCAP]
````

#### Agrupando causa básica de acordo com a tabela de GRUPOS do CID-10

````python
#grupo CID
def agrupamento_cid(row, cids):
    valores = pd.DataFrame()
    valores['CAUSABAS'] = cids.CAUSABAS.str[1:3].astype('int')
    inicio = int(row.CATINIC[1:])
    fim = int(row.CATFIM[1:])
    
    cids.loc[( ( (cids.CAUSABAS.str[0] == row.CATINIC[0]) | (cids.CAUSABAS.str[0] == row.LETRA_FIM) )
              &
               ((cids.CAUSABAS.str[1:3] >= row.CATINIC[1:]) & (cids.CAUSABAS.str[1:3] <= row.CATFIM[1:]))
             )
            , ['COD_GRUPO_CAUSA_BASICA', 'GRUPO_CAUSA_BASICA']] = [row.CATINIC + '-' + row.CATFIM, row.DESCRICAO]
````

#### Atribuindo categoria e subcategorias do CID-10 que a CAUSA BÁSICA equivale

````python
#CATEGORIA CID
#(cids.CAUSABAS.str.len() == 3) &
def categoria_cid(row, cids):
    cids.loc[((cids.CAUSABAS.str[:3] == row.CAT)),
             ['COD_CAUSA_BASICA', 'CAUSA_BASICA']] = [row.CAT, row.DESCRICAO]
````

````python
#SUBCATEGORIA CID 
def sub_categoria_cid(row, cids):
    cids.loc[((cids.CAUSABAS.str.len() == 4) & (cids.CAUSABAS == row.SUBCAT)),
            ['CAUSA_BASICA']] = row.DESCRABREV[6:]
````

#### Função para tratar os valores da coluna CAUSABAS: 
  vai classificar o grupo CID-10, categoria e subcategoria CID-10 e o capítulo CID-10

````python
def nomeando_cids(df):
    df_cid_grupos = pd.read_csv(path + '/CSVs BASE/CID-10-GRUPOS.CSV', header=0, sep=';', encoding='latin1')
    df_cid_grupos.head()

    # df_cid_grupos = df_cid_grupos.drop(columns=['Unnamed: 0'])


    df_cid_cate = pd.read_csv(path + '/CSVs BASE/CID-10-CATEGORIAS.CSV', header=0, sep=';', encoding='latin1')

    df_cid_sub = pd.read_csv(path + '/CSVs BASE/CID-10-SUBCATEGORIAS.CSV', header=0, sep=';', encoding='latin1')
    
    df_cid_cap = pd.read_csv(path + '/CSVs BASE/CID-10-CAPITULOS.CSV', header=0, sep=';', encoding='latin1')

    df_cid_cap['LETRA_INICIO'] = ""
    df_cid_cap['NUM_INICIO'] = ""
    df_cid_cap['LETRA_FIM'] = ""
    df_cid_cap['NUM_FIM'] = ""
    
    df_cid_grupos['LETRA_INICIO'] = ""
    df_cid_grupos['NUM_INICIO'] = ""
    df_cid_grupos['LETRA_FIM'] = ""
    df_cid_grupos['NUM_FIM'] = ""
    
    col_cids = ['CAUSABAS']

    cids = pd.DataFrame(df[col_cids].astype('str'))

#### RETIRANDO O '*' INÍCIO
    cids.apply(limpeza, col_cids=col_cids, axis='columns')

    cids['GRUPO_CAUSA_BASICA'] = ""
    cids['COD_GRUPO_CAUSA_BASICA'] = ""
    cids['COD_CAUSA_BASICA'] = ""
    cids['CAUSA_BASICA'] = ""
    cids['CAPITULO_CAUSA_BASICA'] = ""
    cids['COD_CAPITULO_CAUSA_BASICA'] = ""
    
    # df_cid_cap.apply(capitulo_cid, cids=cids, axis='columns')

    df_cid_grupos.apply(agrupamento_cid, cids=cids, axis='columns')

    df_cid_cate.apply(categoria_cid, cids=cids, axis='columns')
    df_cid_sub.apply(sub_categoria_cid, cids=cids, axis='columns')
    
    return cids
````

#### Tratando a ocupação da pessoa
Criação de duas novas colunas: uma com o código do grupo pertencente e outra com a descrição deste

#####Atribuindo descrição do grupo de ocupação de acordo com o código

````python
def cod_ocup(row, ocupacoes):
    ocupacoes.loc[(ocupacoes.OCUP.str[:6] == row.CODIGO), ['OCUPACAO']] = row.TITULO
````

Atribuindo código do grupo de ocupação de acordo com o código

````python
def cod_group_ocup(row, ocupacoes):
    ocupacoes.loc[(ocupacoes.OCUP.str[0] == row.CODIGO), 
                  ['GRUPO_OCUPACAO', 'CODIGO_GRUPO_OCUPACAO']] = [row.TITULO, row.CODIGO]
````

#### Realizando tratamento das ocupações de acordo com o CSV da Classificação Brasileira de Ocupações

````python
def ocupacoes(df):
#     df_ocupacoes = pd.read_csv('../../Downloads/CBO2002_Ocupacao.csv', header=0, sep=';', encoding='latin1')
    df_grupo = pd.read_csv(path + '/CSVs BASE/CBO2002_Grande_Grupo.csv', header=0, sep=';', encoding='latin1')
    
#     df_ocupacoes['CODIGO'] = df_ocupacoes['CODIGO'].astype('str')
    df_grupo['CODIGO'] = df_grupo['CODIGO'].astype('str')
    
    ocupacoes = pd.DataFrame(df['OCUP'].astype('str'))
#     ocupacoes['OCUPACAO'] = "" 
    ocupacoes['GRUPO_OCUPACAO'] = ""
    ocupacoes['CODIGO_GRUPO_OCUPACAO'] = ""
    
#     df_ocupacoes.apply(cod_ocup, ocupacoes=ocupacoes, axis='columns')
    df_grupo.apply(cod_group_ocup, ocupacoes=ocupacoes, axis='columns')
    ocupacoes['GRUPO_OCUPACAO'] = ocupacoes['GRUPO_OCUPACAO'].fillna('SEM OCUPAÇÃO') 
    
    return ocupacoes['GRUPO_OCUPACAO'], ocupacoes['CODIGO_GRUPO_OCUPACAO'] 
````

#### Criação do CSV com apenas dados em que a causa básica foi "coronavirus" e retirada dessas tuplas do pré-processamento

````python
def retirando_covid(df):
    
    df_covid = df.loc[df['CAUSA_BASICA'].str.contains('coronavirus') == True]
    # df_covid = df_covid.drop(columns=['CAUSA_BASICA'])
    df_covid.columns = df_covid.columns.str.lower()
    df_covid = df_covid.rename(columns={'assistmed': 'teve_assistencia_medica', 'racacor': 'raca_cor', 'lococor': 'local_ocorrencia',
                   'esc2010': 'escolaridade'})
    
    df_covid.to_csv(path + '/Datasets Pre-processados/mortes_por_covid_2019.csv', header=True, sep=';', encoding='latin1', index=False)
    
    df_sem_covid = df.loc[df['CAUSA_BASICA'].str.contains('coronavirus') == False]
    
    return df_sem_covid
````

#### Função principal de limpeza do DATASET
Nela que haverão as chamadas das funções anteriores

````python
def clean_dataset(df):
  
    # retirando as circunstância de óbito que não envolvem doença como causa
    df = df.loc[(df['CIRCOBITO'].isnull()) | (df['CIRCOBITO'] == 9)] 
    
#### DROP COLUMNS
    df = df.drop(columns=['CONTADOR', 'DTOBITO', 'DTNASC', 'HORAOBITO', 'CODESTAB', 'DTATESTADO', 'NUMEROLOTE', 'DTCADASTRO', 'ATESTANTE', 
                          'STCODIFICA', 'CODIFICADO', 'DTOBITO', 'DTNASC']
                )

#### CATEGORIZANDO OS SEXOS

    df['SEXO'] = df['SEXO'].replace([0, 1, 2] , ["I", "M", "F"])

#### CATEGORIZANDO RAÇAS/COR

    df['RACACOR'] = df['RACACOR'].replace([1, 2, 3, 4, 5], 
                                          ['BRANCA', 'PRETA', 'AMARELA', 'PARDA', 'INDIGENA']
                                          )

#### DATA DE NASCIMENTO

#     dt_nasc = clean_date(df, 'DTNASC')


#     df['DIA_NASC'] = dt_nasc['DIA']
#     df['MES_NASC'] = dt_nasc['MES']
#     df['ANO_NASC'] = dt_nasc['ANO']
    
#     df['DIA_NASC'] = df['DIA_NASC'].astype('int')
#     df['MES_NASC'] = df['MES_NASC'].astype('int')
#     df['ANO_NASC'] = df['ANO_NASC'].astype('int')

#### DATA DE ÓBITO

#     dt_obito = clean_date(df, 'DTOBITO')

#     df['DIA_OBITO'] = dt_obito['DIA'].astype('int')
#     df['MES_OBITO'] = dt_obito['MES'].astype('int')
#     df['ANO_OBITO'] = dt_obito['ANO'].astype('int')


    df = df.drop(columns=['IDADEMAE', 'ESCMAE', 'ESCMAE2010', 'OCUPMAE', 'PESO', 'ORIGEM', 'CB_PRE', 
                          'SERIESCMAE', 'GRAVIDEZ', 'GESTACAO', 'SEMAGESTAC', 'PARTO', 'OBITOPARTO', 'OBITOGRAV', 'OBITOPUERP',
                          'ESCMAEAGR1', 'ESCFALAGR1', 'FONTES', 'QTDFILVIVO', 'QTDFILMORT', 'VERSAOSIST', 'VERSAOSCB', 
                          'CODMUNRES', 'NUDIASOBCO', 'DTCADINV', 'DTCONINV', 'TPPOS', 'TPRESGINFO', 'TPMORTEOCO', 'NECROPSIA',
                          'COMUNSVOIM', 'DTINVESTIG', 'DTRECEBIM', 'CAUSAMAT', 'MORTEPARTO', 'ALTCAUSA',
                          'DTCONCASO', 'CIRURGIA', 'ACIDTRAB', 'FONTEINV', 'TIPOBITO', 'ESTCIV', 'ESC', 'SERIESCFAL', 
                          'CODMUNNATU', 'TPOBITOCOR', 'TPNIVELINV', 'NATURAL', 'EXAME', 'FONTES', 'DTCADINF', 'NUDIASOBCO', 
                          'STDONOVA', 'STDOEPIDEM', 'DTRECORIGA', 'DIFDATA','ATESTADO',
                          'LINHAA', 'LINHAB', 'LINHAC', 'LINHAD', 'LINHAII', 'CAUSABAS_O', 'FONTE',
                          # 'TP_ALTERA', 'CB_ALT', 'OPOR_DO', 
                          'CIRCOBITO',
                          'FONTESINF', 'ESTABDESCR', 'NUDIASOBIN', 'NUDIASINF',
                         ]
                )
    
#### AJEITANDO IDADES
    df['IDADE'] = df['IDADE'].fillna(0)

    df['IDADE'] = ajeitando_idades(df)
    
    faixa_etaria(df)
    
    df['IDADE'] = df['IDADE'].astype('int')


#### CATEGORIZANDO OS DADOS

    df['ASSISTMED'] = df['ASSISTMED'].replace([1, 2, 9], 
                                              ['SIM', 'NAO', 'I']
                                             )

    df['LOCOCOR'] = df['LOCOCOR'].replace([1, 2, 3, 4, 5, 9], 
                                          ['HOSPITAL', 'OES', 'DOMICILIO', 'VIAP', 'OUTROS', 'I']
                                         )
    # OES: outros estabelecimentos de saúde
    # I: ignorado
    # VIAP: via pública

    df['ESC2010'] = df['ESC2010'].replace([0, 1, 2, 3, 4, 5, 9], 
                                          ['SEM', 'EF_I', 'EF_II', 'EM', 'SUPERIOR_I', 'SUPERIOR_C', 'I']
                                         )
    # SEM: SEM ESCOLARIDADE
    # EF_I: ensino fundamental I
    # EF_II: ensino fundamental II
    # EM: ensino médio
    # SUPERIOR_I: superior incompleto
    # SUPERIOR_C: superior completo
    # I: ignorado

#### AJEITANDO OS ESTADOS
    df['ESTADO'], df['CODIGO_ESTADO'] = clean_states(df)

#### AJEITANDO CIDs

    cids = nomeando_cids(df)

    df['GRUPO_CAUSA_BASICA'] = cids['GRUPO_CAUSA_BASICA']
    df['COD_GRUPO_CAUSA_BASICA'] = cids['COD_GRUPO_CAUSA_BASICA']
    df['COD_CAUSA_BASICA'] = cids['COD_CAUSA_BASICA']
    df['CAUSA_BASICA'] = cids['CAUSA_BASICA']
    df['CAPITULO_CAUSA_BASICA'] = cids['CAPITULO_CAUSA_BASICA']
    df['COD_CAPITULO_CAUSA_BASICA'] = cids['COD_CAPITULO_CAUSA_BASICA']

    df = df.drop(columns=['CAUSABAS'])
    
#### OCUPACAO DO ÓBITO

    df['OCUP'] = df['OCUP'].astype('str')
    df['GRUPO_OCUPACAO'], df['CODIGO_GRUPO_OCUPACAO']  = ocupacoes(df)
   
    df = df.drop(columns=['OCUP', 'CODMUNOCOR'])
  
    df = retirando_covid(df)
    
    return df
````
