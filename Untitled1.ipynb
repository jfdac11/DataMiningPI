```python
data = pd.read_csv('../../Downloads/dobrano_.csv', header=0, sep=';', low_memory=False, encoding='latin1')
```

```python
df = data[:1000]
```

```python
df['IDADE'] = df['IDADE'].fillna(0)
df['IDADE'] = df['IDADE'].astype('int')
df['IDADE'] = df['IDADE'].astype('str')
```

```python
df.head()
```

```python
df['ANOS_IDADE'] = pd.Series()
```

```python
df.loc[(((df['IDADE'].str.len() == 3) & (df['IDADE'].str[0] != '4') & (df['IDADE'].str[0] != '5')) | df['IDADE'].str.len() < 3), ['ANOS_IDADE']] = '0'
```

```python
df.loc[((df['IDADE'].str.len() == 3) & (df['IDADE'].str[0] == '4') | (df['IDADE'].str[0] == '5')), ['ANOS_IDADE']] = df['IDADE'].str[1:]
```

```python
def convertendo_idade(row):
    unidade = int(row['IDADE'][0])
    anos = 0
    if (unidade >= 4) & (len(row['IDADE']) == 3):
        anos = unidade - 4
        anos += int(row.IDADE[1:])
    row['ANOS_IDADE'] = anos
    print(row.ANOS_IDADE)
```

```python
df.apply(convertendo_idade, axis='columns')
```

```python
df['IDADE'] = df['ANOS_IDADE']
df = df.drop(columns='ANOS_IDADE')
```
